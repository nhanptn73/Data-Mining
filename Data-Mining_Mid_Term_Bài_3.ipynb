```python
def crawlData():
  #Click chọn Kết quả hoạt động kinh doanh để xem
  business_results = driver.find_element(By.XPATH, "/html/body/form/div[3]/div/div/div/div[2]/div[2]/div[4]/div/div/div/a")
  business_results.click()
  sleep(2)
  #Click chọn 4 năm gần nhất
  near4y = driver.find_element(By.XPATH, "/html/body/form/div[3]/div/div/div/div[2]/div[2]/div[9]/table[1]/tbody/tr/td[2]/span[2]/input")
  near4y.click()
  sleep(2)
  html = driver.page_source
  soup = BeautifulSoup(html,'lxml')
  return soup
```

```python
table = crawlData().find("table",id="tableContent")
table
```

```python
def rowName():
  r_names = []
  t =  crawlData().findAll("td",class_ = "h_t")
  for i in range(len(t)):
    r_n =  t[i].text.strip(' \t\n\r')
    r_names.insert(0,r_n)
  r_names.reverse()
  return r_names
```

```python
import numpy as np
import matplotlib.pyplot as plt
def addlabels(x,y,s):
  for i in range(len(x)):
    plt.text(i, y[i], y[i], ha = s,)

plt.figure(figsize=(12,8))
ax1 = plt.subplot(1,1,1)
ax1.bar(data.index,data.Data2,color= '#63a1db',label='Doanh thu')
addlabels(data.index,data.Data2,'center')
ax1.set_ylim(0)
plt.ylim(0,160000)
ax2=ax1.twinx()
ax2.plot(data.index,data.Data10,'o-',color="#ffc724", label='Lợi nhuận sau thuế')
addlabels(data.index,data.Data10,'left')
plt.ylim(0,40000)
ax1.legend(loc="upper left",bbox_to_anchor=(0,0.95))
ax2.legend(loc="upper left", bbox_to_anchor=(0,1))
plt.title('Kết quả kinh doanh Hòa Phát\n  (Đvt: tỷ đồng)',fontsize= 24,weight='bold', color="#07294d")
plt.grid()
plt.show()
```
